# üßπ <span style="color: DarkGreen; font-family:Times New Roman;">Preprocessing</span> <a name="overview"></a>
---


```python
target_columns = ["Age","Annual Income (k$)", "Spending Score (1-100)"]
```

```python
fig_before_clustering = px.scatter_3d(df, x = target_columns[0], y = target_columns[1], z = target_columns[2], color_continuous_scale=px.colors.sequential.Rainbow)
fig_before_clustering.update_layout(title_text='Supermarket Customer Segmentation Analysis (Before Clustering)', title_x=0.5)
fig_before_clustering.write_html("fig_before_clustering.html")
fig_before_clustering.write_image("fig_before_clustering.png")
fig_before_clustering.show()
```

* * *
# ü´ß <span style="color: DarkGreen; font-family:Times New Roman;">K-Means Clustering</span> <a name="overview"></a>
---


```python
# Define Random State
RS = 13
```

```python
# KMeans clustering using deafult values: n_clusters=8, max_iter=300 
kmeans_default = KMeans(random_state = RS)
kmeans_default.fit(df[target_columns])
```

```python
df_kmeans_default = df.copy()
df_kmeans_default["cluster"] = kmeans_default.labels_
```

```python
fig_kmeans_default = px.scatter_3d(df_kmeans_default, x = target_columns[0], y = target_columns[1], z = target_columns[2], color = "cluster", hover_name = "CustomerID", color_continuous_scale=px.colors.sequential.Rainbow)
fig_kmeans_default.update_layout(title_text='Supermarket Customer Segmentation Analysis (Clustering KMeans Default, K=8)', title_x=0.5)
fig_kmeans_default.write_html("fig_kmeans_default.html")
fig_kmeans_default.show()
```

* * *
# üßë‚Äçüî¨ <span style="color: DarkGreen; font-family:Times New Roman;">Tuning</span> <a name="model-selection"></a>
---


```python
distortions = []
K = range(1, 10)

for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state = RS)
    kmeanModel.fit(df[target_columns])

    cdist_result = cdist(df[target_columns], kmeanModel.cluster_centers_,'euclidean')
    minimum_cdist = np.min(cdist_result, axis = 1)
    sum_minimum_cdsit = np.sum(minimum_cdist)
    avg_cdist = sum_minimum_cdsit / df.shape[0]

    distortions.append(avg_cdist)
```

```python
plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
elbow = pd.DataFrame({
    "k" : K,
    "distortions" : distortions
})
```

```python
elbow["growth_rate"] = elbow.distortions.pct_change()
```

```python
elbow
```

```python
best_k = 5
```

* * *
# üß´ <span style="color: DarkGreen; font-family:Times New Roman;">Final Clustering</span> <a name="final-evaluation"></a>
---


```python
# KMeans clustering using deafult values: n_clusters=8, max_iter=300 
kmeans_final = KMeans(n_clusters = best_k,random_state = RS)
kmeans_final.fit(df[target_columns])
```

```python
df_kmeans_final = df.copy()
df_kmeans_final["cluster"] = kmeans_final.labels_ + 1
```

```python
fig_kmeans_final = px.scatter_3d(df_kmeans_final, x = target_columns[0], y = target_columns[1], z = target_columns[2], color = "cluster", hover_name = "CustomerID", color_continuous_scale=px.colors.sequential.Rainbow)
fig_kmeans_final.update_layout(title_text='Supermarket Customer Segmentation Analysis (Clustering KMeans, K=5)', title_x=0.5)
fig_kmeans_final.write_html("fig_kmeans_final.html")
fig_kmeans_final.write_image("fig_kmeans_final.png")
fig_kmeans_final.show()
```
